```python
# import necessary libraries
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
university_df.dropna(inplace = True)
university_df.isnull().count()
```

```python
university_df.drop(['link', 'logo', 'score'], axis=1, inplace=True) 
university_df['research_output'] = university_df['research_output'].replace('Very high', 'Very High')
university_df['international_students'] = university_df['international_students'].apply(lambda x: float(str(x).replace(',','')))
university_df['faculty_count'] = university_df['faculty_count'].apply(lambda x: float(str(x).replace(',','')))
```

```python
year_df = university_df['year'].value_counts().sort_values()
fig, ax = plt.subplots(dpi=90)
sns.countplot(data=university_df, x='year')
ax.bar_label(ax.containers[0])

plt.xlabel('Year')
plt.ylabel('Number of Universities')
plt.title('Number of Universities Ranked per Year')
```

```python
type_df = university_df['type'].value_counts()
myexplode = [0,0.2]
plt.pie(university_df['type'].value_counts().values, labels = university_df['type'].value_counts().index,autopct='%.1f%%', shadow = 'true', explode=myexplode)
plt.title('Types of University')
```

```python
plt.figure(figsize=(10, 4)) 
c = ['red', 'yellow', 'black', 'blue', 'green']
university_df['region'] = university_df['region'].apply(lambda x: x.replace('Latin America', 'South America'))
region_sum = pd.DataFrame(university_df['region'].value_counts().reset_index())
ax = plt.bar(region_sum['index'], region_sum['region'])

plt.title('Number of Universities per Continent')
plt.xlabel('Continent')
plt.ylabel('Number of Universities') 


```

```python
uni_df = university_df['university'].value_counts()

fig, ax = plt.subplots(figsize=(10,18), dpi=150)

sns.countplot(data=university_df, y='country', order=university_df.country.value_counts().index);
plt.xlabel('Number of Universities')
plt.ylabel('Country')
plt.title("Distribution of Universities across Countries")
```

```python
top_unis = ['Massachusetts Institute of Technology (MIT) ', 'Stanford University', 'University of Oxford', 'Harvard University', 'University of Cambridge', 'California Institute of Technology (Caltech)', 'ETH Zurich - Swiss Federal Institute of Technology', 'Imperial College London', 'UCL', 'University of Chicago']
topunis_df = university_df[university_df['university'].isin(top_unis)][['year','university','rank_display']].reset_index(drop=True)

fig = plt.figure(figsize=(15,15), dpi=100)

for uni, i in zip(top_unis, range(1, 11)):
    new_df = topunis_df[topunis_df['university'] == uni]
    ax = fig.add_subplot(5, 2, i)
    ax.plot(new_df['year'], new_df['rank_display'])
    plt.gca().invert_yaxis()
    ax.set_title(uni)
    
    
fig.subplots_adjust(wspace=0.2, hspace=0.6, top=0.92)
fig.suptitle('Ranking of top 10 universities from 2017 to 2022');
```

```python
fig, ax = plt.subplots(figsize=(8,6), dpi=90)

sns.countplot(data=university_df, x='research_output', hue='type')

for container in ax.containers:
    ax.bar_label(container)

plt.ylabel('Number of Universities')
ax.set_xlabel('Research Output')
fig.suptitle('Research Output of Universities')
```

```python
fig, ax = plt.subplots(figsize=(8,5), dpi=90)

sns.barplot(data=university_df, x='research_output', y='faculty_count', hue='type', ci=None)

ax.set_xlabel('Research Output')
ax.set_ylabel('Faculty Count')
ax.legend(loc="upper right")

fig.suptitle('Research Output Vs Faculty Count');
```

```python
intstu_country = pd.DataFrame(university_df.groupby(['country'], sort=False)['international_students'].sum().sort_values(ascending=False)[:10])

fig, ax = plt.subplots(figsize=(10,4), dpi=100)

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.tick_params(left=False)
ax.get_xaxis().set_visible(False)

sns.barplot(data=intstu_country, x='international_students', y=intstu_country.index);

ax.bar_label(ax.containers[0], fmt = '%d')

ax.set_ylabel('Country');
fig.suptitle('Country of Choice for International Students from 2017 - 2022');
```

```python
from sklearn.preprocessing import LabelEncoder
LE = LabelEncoder()
university_df['country'] = LE.fit_transform(university_df['country'])
university_df['region'] = LE.fit_transform(university_df['region'])
university_df['type'] = LE.fit_transform(university_df['type'])
university_df['research_output'] = LE.fit_transform(university_df['research_output'])
university_df['size'] = LE.fit_transform(university_df['size'])
university_df['year'] = LE.fit_transform(university_df['year'])
university_df['city'] = LE.fit_transform(university_df['city'])
university_df['rank_display'] = LE.fit_transform(university_df['rank_display'])
university_df['university'] = LE.fit_transform(university_df['university'])


university_df.head()
```

Data Modelling

```python
x = university_df.drop(['rank_display'] , axis = 1).values
y = university_df['rank_display'].values
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
sc.fit_transform(university_df)
```

Data Splitting

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state = 0)
```

```python
from sklearn.metrics import r2_score
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train,y_train)
y_pred = lr.predict(x_test)

plt.scatter(y_test,y_pred)
r2_score(y_test,y_pred)
```

```python
from sklearn.tree import DecisionTreeRegressor
dtr = DecisionTreeRegressor()
dtr.fit(x_train,y_train)
y_pred = dtr.predict(x_test)

plt.scatter(y_test,y_pred)
r2_score(y_test,y_pred)
```

```python
from sklearn.ensemble import RandomForestRegressor
rfr = RandomForestRegressor()
rfr.fit(x_train,y_train)
y_pred = rfr.predict(x_test)

plt.scatter(y_test,y_pred)
r2_score(y_test,y_pred)
```
